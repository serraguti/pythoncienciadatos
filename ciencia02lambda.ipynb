Declaramos un conjunto de números

```python
listaNumeros = [2,6,33,22,33,44,1]
listaNumeros
```

Lambda nos permite ejecutar tareas complejas, pero solo
con una línea
Quiero multiplicar cada número de la Listas x 2 y 
ver su doble
Para recorrer cada elemento en lambda, tenemos map


```python
listaDoble = list(map(lambda x:x*2, listaNumeros))
listaDoble
```

Otra utilidad radica en realizar filtros sobre un cojunto.
Para ello, se utiliza filter de lambda
Por ejemplo, podemos recuperar los números positivos de una lista

```python
lista = [2, 4, -5, -788, 22, 1, -99]
listaPositivos = list(filter(lambda x:x > 0, lista))
listaPositivos
```
