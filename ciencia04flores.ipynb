Vamos a averiguar qué podemos hacer si tuvieramos valores nulos
dentro de nuestros datos.
Para trabajar con estos valores se utiliza la librería
numpy

```python
import numpy as np
```

Incluimos valores nulos en las tres primeras filas: np.nan

```python
df.loc[:8, 'longitud_petalo'] = np.nan
df.head()
```

```python
df.info()
```

Los datos nulos son elementos de lo más normal dentro de un DataFrame, datos que no se han rellenado.
Para tratar datos nulos, tenemosun par de opciones
dropna() --> Elimina la columna/filas con nulos
Podemos indicar axis=0 para filas
axis = 1 para las columnas

```python
df_limpio = df.dropna(axis=1)
df_limpio.head()
```

```python
df = df.dropna(axis=0)
df
```

La otra posibilidad está en poder operar con los datos 
dentro del DataFrame.  Un Null es lo mismo aquí que en base de 
datos, no se tiene en cuenta.
En lugar de limpiar el DataFrame de nulos, podemos modificar los valores nulos también.
Es decir, quitar el null y poner un valor por defecto
Para ello se utiliza fillna(valor)

```python
df['longitud_petalo'] = df['longitud_petalo'].fillna(999)
df
```

También podemos crear agrupaciones para averiguar características respecto a una columna.  Esto es lo mismo que un GROUP BY dentro de SQL Server.

```python
df_grupo = df.groupby('clase')['ancho_petalo'].mean()
df_grupo
```

Por supuesto, podemos realizar hasta dibujos!!!!!
