```python
import pandas as pd
```

Creamos un diccionario con paises

```python
dataPaises = {
    "pais": ["España", "Francia", "Colombia", "Mexico", "Peru"],
    "capital": ["Madrid", "Paris", "Bogota", "Ciudad de Mexico", "Lima"],
    "superficie": [1.9, 2.6, 3.9, 4.0, 2.8] 
}
```

Creamos un Dataframe con el diccionario

```python
df = pd.DataFrame(dataPaises)
df
```

Por ejemplo, podemos dibujar solamente una columna

```python
df[['capital']]
```

Podemos dibujar más de una columna

```python
df[['capital', 'superficie']]
```

Se pueden seleccionar filas por su posición del índice
Primero incluido y el segundo no incluido

```python
df[2:5]
```

Por defecto, los índices de los elementos son numéricos y 
los pone automáticamente.  
Podemos crear nuestros propios indices y sustituir el index

```python
df.index = ["ES", "FR", "CO", "MX", "PE"]
df = df.reset_index()
df
```
